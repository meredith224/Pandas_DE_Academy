```python

import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Set seed for reproducibility
np.random.seed(42)

# Generate base data
n_records = 150

# Create messy sales data
data = {
    'order_id': range(1001, 1001 + n_records),
    'customer_name': np.random.choice(
        ['John Smith', 'jane doe', 'MARY JOHNSON', 'Bob Williams',
         'Alice Brown', 'charlie Davis', None, 'N/A', ''],
        n_records
    ),
    'email': [
        f'customer{i}@email.com' if np.random.random() > 0.1 else
        np.random.choice(['INVALID', None, '', 'customer@', '@email.com'])
        for i in range(n_records)
    ],
    'order_date': [
        (datetime(2024, 1, 1) + timedelta(days=int(np.random.exponential(30)))).strftime('%Y-%m-%d')
        if np.random.random() > 0.05 else
        np.random.choice(['2024-13-45', '01/15/2024', 'Invalid', None, ''])
        for _ in range(n_records)
    ],
    'product': np.random.choice(
        ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'laptop', 'MOUSE',
         'Headphones', None, 'Unknown', ''],
        n_records
    ),
    'quantity': [
        int(np.random.randint(1, 10)) if np.random.random() > 0.08 else
        np.random.choice([-5, 0, 999, None, 'three'])
        for _ in range(n_records)
    ],
    'price': [
        round(np.random.uniform(10, 1000), 2) if np.random.random() > 0.07 else
        np.random.choice([-50, 0, 99999.99, None, 'FREE'])
        for _ in range(n_records)
    ],
    'region': np.random.choice(
        ['North', 'South', 'East', 'West', 'north', 'SOUTH',
         None, 'N/A', '', 'Unknown'],
        n_records
    ),
    'status': np.random.choice(
        ['Shipped', 'Pending', 'Delivered', 'shipped', 'CANCELLED',
         None, '', 'In Transit', 'Unknown'],
        n_records
    )
}

# Create DataFrame
df = pd.DataFrame(data)

# Add some duplicate rows
df = pd.concat([df, df.sample(10)], ignore_index=True)

# Add extra whitespace to some string columns
mask = np.random.random(len(df)) > 0.9
df.loc[mask, 'customer_name'] = df.loc[mask, 'customer_name'].apply(
    lambda x: f'  {x}  ' if pd.notna(x) and x != '' else x
)

# Export to CSV
df.to_csv('messy_sales_data.csv', index=False)
print(f"Generated messy_sales_data.csv with {len(df)} records")
print(f"File created successfully!")
```
