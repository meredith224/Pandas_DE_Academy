```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
num_rows = 2000

data = {
    'Day': range(1, num_rows + 1),
    'Product': np.random.choice(['Apples', 'Bananas', 'Cherries', 'Dates'], num_rows),
    'Price': np.random.randint(10, 100, num_rows),
    'Units_Sold': np.random.randint(1, 20, num_rows)
}

df = pd.DataFrame(data)

# Calculate a Total Revenue column
df['Revenue'] = df['Price'] * df['Units_Sold']

print("First 5 rows of data:")
print(df.head())
print("-" * 30)
```

```python
# A. Histogram: Show the distribution of Revenue
# "How common are different revenue amounts?"
df['Revenue'].plot(kind='hist', bins=20, title='Distribution of Revenue', color='skyblue', edgecolor='black')
plt.xlabel('Revenue Amount')
plt.show()
```

```python
# B. Bar Chart: Average Revenue per Product
# "Which product makes the most money on average?"
avg_revenue = df.groupby('Product')['Revenue'].mean()
print(avg_revenue)
avg_revenue.plot(kind='bar', title='Average Revenue by Product', color='red')
plt.ylabel('Average Revenue')
plt.show()
```

```python
# C. Scatter Plot: Price vs Units Sold
# "Does a higher price mean we sell fewer units?"
df.plot(kind='scatter', x='Price', y='Units_Sold', title='Price vs. Units Sold', color='green', alpha=.5)
plt.show()
```

```python
# D. Line Plot: Revenue over "Time" (Day)
# "How did revenue change over the first 50 days?"
# We limit to 50 just so the chart isn't too crowded
subset = df.head(50)
subset.plot(kind='line', x='Day', y='Revenue', title='Revenue Over First 50 Days')
plt.ylabel('Revenue')
plt.show()
```

```python
# E. Horizontal Bar Chart: Total Units Sold by Product
# "Which product has sold the most units total?"
plt.figure()
total_units = df.groupby('Product')['Units_Sold'].sum()
total_units.plot(kind='barh', title='Total Units Sold by Product', color='purple')
plt.xlabel('Total Units')
plt.show()
```

```python
# F. Pie Chart: Market Share (Revenue)
# "What % of our total money comes from each product?"
plt.figure()
total_revenue = df.groupby('Product')['Revenue'].sum()
total_revenue.plot(kind='pie', title='Revenue Share by Product', autopct='%1.1f%%')
plt.ylabel('') # Hides the 'Revenue' label on the side
plt.show()
```

```python
# G. Box Plot: Price Ranges
# "What is the range of prices? Are there any outliers?"
plt.figure()
df['Revenue'].plot(kind='box', title='Box Plot of Prices')
plt.show()

print(df)
```
